# Similarity

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
car =       [27,4,24]
auto =      [3,33,0]
insurance = [0,33,29]
best =      [14,0,17]

m = np.array([car, auto, insurance, best])
```

```python
print(m[0,:])
print(m[:,0])
```

# Sim function

```python
def Sim(q, d):
    p = 0; r1 = 0; r2 = 0
    n = len(q)
    for i in range(n):
        p += q[i]*d[i]
        r1 += q[i]**2 
        r2 += d[i]**2       
    return p / (r1**0.5 * r2**0.5)
```

# q vector

```python
q = [13,0,0,22]
```

```python
for i in range(3):
    print("q vs doc"+str(i+1)+" => "+str(Sim(q, m[:,i])))
```

```python
mm = np.array([[27,4,24,13], [3,33,0,0], [0,33,29,0], [14,0,17,22]])
```

```python
plt.quiver(mm[0,:], mm[1,:], mm[2,:], mm[3,:], angles='xy', scale_units='xy', color=['r','b','g','m'], scale=1)
plt.xlim(-1, 58)
plt.ylim(-1, 58)
plt.xlabel('car')
plt.ylabel('best')

plt.show()
```

## Focusing only on "car" and "best"

```python
q2 = [13,22]
m2 = np.array([[27,14],
                [4,0],
                [24,17]])
```
